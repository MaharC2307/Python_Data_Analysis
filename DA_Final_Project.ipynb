```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 12))
fig.suptitle('Histograms of columns', fontsize=16)

alpha_col = ['job', 'marital', 'education', 'housing', 'loan', 'contact','month', 'day_of_week', 'poutcome']

for i, col in enumerate(alpha_col):
    sns.histplot(df, x=col, kde=True, hue='Subscribed', ax=axes[i//3, i%3], multiple="stack")
    axes[i//3, i%3].set_title(col)
    axes[i//3, i%3].tick_params(axis='x', rotation=90)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
custom_palette = {'yes': 'green', 'no': 'red'}

# Plot histogram
sns.histplot(df, kde=True, x='job', hue='Subscribed', palette=custom_palette)

# Rotate x-axis labels
plt.xticks(rotation=90)

plt.show()
```
